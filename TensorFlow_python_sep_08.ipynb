## TensorFlow

TensorFlow as a Homework Helper

**Homework Questions = Input Data**

Imagine you have a bunch of math questions: 2 + 3 = ?, 5 × 6 = ?.

These are like the data you give to TensorFlow.

**Calculator = Operations (ops)**

To solve each question, you need tools: add, subtract, multiply, divide.

TensorFlow has all these tools built-in.

**Answer Sheet = Tensors**

When you write down answers, they go into boxes.

TensorFlow also stores numbers in boxes called tensors.

**Teacher’s Key = Labels**

Your teacher already has the correct answers.

TensorFlow uses these answers to check how well it did.

**Grading = Loss Function**

If you got 3 wrong, your grade shows how far you are from perfect.

TensorFlow measures this “wrongness” using a loss function.

**Learning from Mistakes = Training**

After checking, you fix your mistakes and try again.

TensorFlow does the same by adjusting its weights (the way it solves problems).

**Practicing Over Time = Iterations/Epochs**

Each day you practice more homework, you get better.

TensorFlow also improves by repeating many times until it learns.

**Becoming Fast at Homework = Model Prediction**

After practice, you can answer new questions quickly without help.

That’s TensorFlow making predictions once it’s trained.

## 5. Loss Function

Measures how wrong the model is compared to labels.
Common: Mean Squared Error (MSE) for regression.

```python
# Suppose model predicts [2.1, 3.9, 5.8, 8.2, 9.7]
preds = np.array([2.1, 3.9, 5.8, 8.2, 9.7])
loss = tf.reduce_mean(tf.square(y-preds))
print("Loss:", loss.numpy())
```

## 6. Training

Process of adjusting model weights to reduce loss.
In Python/TensorFlow, we use optimizers like SGD or Adam.

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[1])
])
model.compile(optimizer='sgd', loss='mean_squared_error')
```

## 7. Iterations/Epochs

How many times the model sees the dataset while training.

Iteration = one update step.

Epoch = one pass through the entire dataset.

```python
history = model.fit(X, y, epochs=5, verbose=0)  # 500 passes
```

## 9. Full implementation of Tensor Flow

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# 1. Input Data (features and labels)
X = np.array([1, 2, 3, 4, 5], dtype=float)   # hours studied
y = np.array([2, 4, 6, 8, 10], dtype=float)  # exam scores (double)

# 2. Define Model (y = Wx + b)
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[1])
])

# 3. Compile Model (Optimizer + Loss Function)
model.compile(optimizer='sgd', loss='mean_squared_error')

# 4. Train Model (Epochs = passes through data)
history = model.fit(X, y, epochs=200, verbose=0)

# 5. Predict on new data
prediction = model.predict(np.array([6]))
print("Predicted score for 6 hours study:", prediction[0][0])

# 6. Plot Loss
plt.plot(history.history['loss'])
plt.xlabel("Epochs")
plt.ylabel("Loss (MSE)")
plt.title("Training Loss over Time")
plt.show()
```

## 10. Full implemenatation of PyTorch

```python
import torch
import torch.nn as nn

# True labels (y) and predictions (preds)
y = torch.tensor([2, 4, 6, 8, 10], dtype=torch.float32)
preds = torch.tensor([2.1, 3.9, 5.8, 8.2, 9.7], dtype=torch.float32)

# Define Mean Squared Error Loss function
mse_loss = nn.MSELoss()

# Compute loss
loss = mse_loss(preds, y)
print("MSE Loss:", loss.item())
```
